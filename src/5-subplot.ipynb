### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line chart
x_line = np.arange(0, 10, 0.1)
y_line = np.sin(x_line)

# Data for bar chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Data for scatter plot
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart
axs[0, 0].plot(x_line, y_line)
axs[0, 0].set_title('Line Chart')

# Bar chart
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
